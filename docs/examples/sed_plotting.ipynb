# SED Plotting

With HostPhot, the user can plot the SED from the calculated photometry.

```python
import hostphot
print('HostPhot version:', hostphot.__version__)
```

```python
from hostphot.sed_plotting import plot_sed
name = 'SN2004eo'
phot_type = 'global'
z = 0.0157

plot_sed(name, phot_type, z=z)
```
